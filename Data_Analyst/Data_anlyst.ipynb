```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Đảm bảo các cột không bị lỗi kiểu dữ liệu
df['A3. Độ tuổi'] = df['A3. Độ tuổi'].astype(str)

# === 1. Độ tuổi vs Mức độ quen thuộc với AI ===
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='A3. Độ tuổi', y='A6. Mức độ quen thuộc với AI nói chung', palette='coolwarm')
plt.title('Mức độ quen thuộc với AI theo độ tuổi')
plt.xlabel('Nhóm tuổi')
plt.ylabel('Mức độ quen thuộc')
plt.grid(alpha=0.3)
plt.show()

# === 2. Độ tuổi vs Việc đã sử dụng DeepSeek ===
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='A3. Độ tuổi', hue='A7. Bạn đã từng sử dụng Deepseek chưa?', palette='Set2')
plt.title('Tỷ lệ sử dụng DeepSeek theo độ tuổi')
plt.xlabel('Nhóm tuổi')
plt.ylabel('Số lượng người tham gia')
plt.legend(title='Đã sử dụng DeepSeek chưa?')
plt.grid(alpha=0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Tạo bảng tần suất
ai_familiarity = df['A6. Mức độ quen thuộc với AI nói chung'].value_counts()

# Thiết lập style seaborn
sns.set(style="whitegrid")

# Tạo bảng màu pastel nhạt
colors = sns.color_palette("pastel", n_colors=len(ai_familiarity))

# Vẽ biểu đồ tròn
plt.figure(figsize=(7,7))
wedges, texts, autotexts = plt.pie(
    ai_familiarity.values,
    labels=None,  # bỏ label trên pie, sẽ dùng legend
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    textprops={'fontsize': 11}
)

# Thêm legend (chú thích)
plt.legend(
    wedges,
    ai_familiarity.index,
    title="Mức độ quen thuộc",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1),
    fontsize=11
)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Tạo bảng tần suất cho câu hỏi A7
deepseek_usage = df['A7. Bạn đã từng sử dụng Deepseek chưa?'].value_counts()

# Thiết lập style seaborn
sns.set(style="whitegrid")

# Tạo bảng màu pastel nhạt với độ trong suốt nhẹ
colors = sns.color_palette("pastel", n_colors=len(deepseek_usage))
colors = [(r, g, b, 0.6) for r, g, b in colors]  # làm màu nhạt hơn nữa

# Vẽ biểu đồ tròn
plt.figure(figsize=(7,7))
wedges, texts, autotexts = plt.pie(
    deepseek_usage.values,
    labels=None,  # ẩn nhãn, dùng legend thay
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    textprops={'fontsize': 11}
)

# Thêm legend (chú thích)
plt.legend(
    wedges,
    deepseek_usage.index,
    title="Mức độ sử dụng DeepSeek",
    loc="center left",
    bbox_to_anchor=(1, 0, 0.5, 1),
    fontsize=11
)

plt.title('Bạn đã từng sử dụng DeepSeek chưa?', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
